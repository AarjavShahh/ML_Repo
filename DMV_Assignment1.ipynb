```python
#Q19. Image manipulation
import numpy as np

grayscale_image = np.random.randint(0, 256, size=(5, 5))
inverted_image = 255 - grayscale_image
thresholded_image = np.where(grayscale_image > 128, 255, 0)
print("Grayscale Image: \n", grayscale_image)
print("Inverted Image: \n", inverted_image)
print("Thresholded Image: \n", thresholded_image)
```
