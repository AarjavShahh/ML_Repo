```python
import numpy as np
import pandas as pd

np.random.seed(123)

n_samples = 1000
class_0_ratio = 0.7
c0 = int(n_samples * class_0_ratio)
c1 = n_samples - c0
```

```python
c0, c1
```

```python
c0 = pd.DataFrame({
    'feature_1': np.random.normal(loc=0, scale=1, size=c0),
    'feature_2': np.random.normal(loc=0, scale=1, size=c0),
    'target': [0] * c0
})

class_1 = pd.DataFrame({
    'feature_1': np.random.normal(loc=2, scale=1, size=c1),
    'feature_2': np.random.normal(loc=2, scale=1, size=c1),
    'target': [1] * c1
})
```

```python
df=pd.concat([c0, class_1]).reset_index(drop=True)
```

```python
df.tail(20)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import mean_squared_error, log_loss

X = df[['feature_1', 'feature_2']]
y = df['target']

model = LogisticRegression(random_state=42, solver='liblinear')
model.fit(X, y)

print("Logistic Regression model trained successfully.")
```

```python
y_pred_proba = model.predict_proba(X)[:, 1]

print("Predicted probabilities (first 5):\n", y_pred_proba[:5])
```

```python
mse_dataset = mean_squared_error(y, y_pred_proba)
print(f"Mean Squared Error (MSE) for the dataset: {mse_dataset:.4f}")

bce_dataset = log_loss(y, y_pred_proba, labels=[0, 1])
print(f"Binary Cross-Entropy (BCE) for the dataset: {bce_dataset:.4f}")
```
