```python
#load iris dataset
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.datasets import load_iris
iris = load_iris()
iris
```

```python
df = pd.DataFrame(iris.data, columns=iris.feature_names)
df
```

```python
df['target'] = iris['target'] #'setosa' - 0, 'versicolor' - 1, 'virginica' - 2
```

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, classification_report
```

```python
# Define features (X) and target (y)
X = df.drop('target', axis=1)
y = df['target']
```

```python
# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
# Initialize and train the Random Forest Classifier
model = RandomForestClassifier(n_estimators=100, random_state=42)
model.fit(X_train, y_train)
```

```python
# Make predictions on the test set
y_pred = model.predict(X_test)

# Evaluate the model
print(f"Accuracy: {accuracy_score(y_test, y_pred):.2f}")
print("\nClassification Report:")
print(classification_report(y_test, y_pred))
```

```python
new_sample_data = pd.DataFrame({
    'sepal length (cm)': [5.2, 6.0, 7.0],
    'sepal width (cm)': [2.1, 2.7, 3.2],
    'petal length (cm)': [3.8, 4.0, 6.1],
    'petal width (cm)': [1.2, 1.2, 2.0]
})

print("New sample data created:")
print(new_sample_data)
```

**Reasoning**:
Now that the new sample data is defined, the next logical step is to use the trained `RandomForestClassifier` model to predict the class labels for this data.



```python
new_predictions = model.predict(new_sample_data)

# Map numerical predictions to actual class names for better readability
predicted_species = [iris.target_names[p] for p in new_predictions]

print("Predicted class for new sample data:")
for i, prediction in enumerate(predicted_species):
    print(f"Sample {i+1}: {prediction}")
```

## Explain Precision, Recall, and F1-score

### Subtask:
Elaborate on precision, recall, and F1-score for each class (setosa, versicolor, virginica). Precision is the ratio of correctly predicted positive observations to the total predicted positives. Recall (Sensitivity) is the ratio of correctly predicted positive observations to all observations in actual class. F1-score is the weighted average of Precision and Recall.


### Explanation of Precision, Recall, and F1-score

- **Precision**: For a given class, precision is the ratio of correctly predicted positive observations to the total predicted positives. It answers the question: "Of all the instances predicted as positive for this class, how many were actually positive?"

- **Recall (Sensitivity)**: For a given class, recall is the ratio of correctly predicted positive observations to all actual observations in that class. It answers the question: "Of all the instances that were actually positive for this class, how many did we correctly predict?"

- **F1-score**: The F1-score is the harmonic mean of Precision and Recall. It provides a single metric that balances both precision and recall, making it useful when you need to consider both false positives and false negatives. It is particularly helpful when the class distribution is imbalanced.

## Explain Averages

### Subtask:
Explain 'macro avg' and 'weighted avg'. Macro average calculates metrics independently for each class and then takes the average (unweighted). Weighted average accounts for class imbalance by weighting the average of each class by its support.


### Explanation of Macro Average and Weighted Average

-   **Macro Average**: This metric calculates the average of precision, recall, and F1-score independently for each class and then takes the unweighted mean. It treats all classes equally, regardless of how many instances belong to each class. If there's an imbalance in the number of samples per class, the macro average will give equal importance to the performance on the minority classes as it does to the majority classes.

-   **Weighted Average**: This metric calculates the average of precision, recall, and F1-score for each class, but it weights each class's metric by its `support` (the number of true instances for that class). This means that classes with more samples contribute more to the overall average. The weighted average is particularly useful when dealing with imbalanced datasets, as it provides a more realistic view of the model's performance on the dataset as a whole.
