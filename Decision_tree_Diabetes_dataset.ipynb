```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
from sklearn.datasets import load_iris
```

```python
iris = load_iris()
```

```python
iris
```

```python
print(iris['DESCR'])
```

```python
X = pd.DataFrame(iris['data'], columns=['sepal length in cm', 'sepal width in cm', 'petal length in cm', 'petal width in cm'])
```

```python
Y = iris['target']
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=10)
```

```python
#apply decision tree classifier
from sklearn.tree import DecisionTreeClassifier
dt = DecisionTreeClassifier(max_depth=2)
#dt = DecisionTreeClassifier()
dt.fit(X_train, Y_train)
```

```python
#Visualize dt

from sklearn import tree
plt.figure(figsize=(10,10))
tree.plot_tree(dt, filled=True)
plt.show()
```

```python
Y_pred=dt.predict(X_test)
```

```python
from sklearn.metrics import confusion_matrix, classification_report
cm = confusion_matrix(Y_test, Y_pred)
print(cm)
print(classification_report(Y_test, Y_pred))
```

```python
from sklearn.datasets import load_diabetes
```

```python
df = load_diabetes()
print(df['DESCR'])
```

```python
df['target']
```

```python
X = pd.DataFrame(df['data'], columns=['age', 'sex', 'bmi', 'bp', 's1', 's2', 's3', 's4', 's5', 's6'])
```

```python
X.head()
```

```python
Y = df['target']
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.3, random_state=10)
```

```python
X_train.head()
```

```python
X_train.corr()
```

```python
from sklearn.tree import DecisionTreeRegressor
reg = DecisionTreeRegressor()
reg.fit(X_train, Y_train)
```

```python
param = {
    'criterion':['squared_error','friedman_mse','absolute_error'],
    'splitter':['best','random'],
    'max_depth':[1,2,3,4,5,10, 15, 20, 25],
    'max_features':['auto','sqrt','log2']
}
```

```python
param
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
reg = DecisionTreeRegressor()
```

```python
grid = GridSearchCV(reg, param, cv=5, scoring='neg_mean_squared_error')
```

```python
import warnings
warnings.filterwarnings('ignore')
grid.fit(X_train, Y_train)
```

```python
grid.best_params_
```

```python
Y_pred = grid.predict(X_test)
```

```python
from sklearn.metrics import mean_squared_error, r2_score, mean_absolute_error
print(r2_score(Y_test, Y_pred))
print(mean_absolute_error(Y_test, Y_pred))
print(mean_squared_error(Y_test, Y_pred))
```

```python
selectedmodel = DecisionTreeRegressor(criterion='friedman_mse', max_depth=5, max_features='log2', splitter='random')
selectedmodel.fit(X_train, Y_train)
```
