```python
df['species'] = df['species'].map({'setosa': 0, 'versicolor': 1, 'virginica': 2})
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
```

```python
X = df.drop('species', axis=1)
y = df['species']
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=42)
```

```python
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

```python
from sklearn.cluster import KMeans
```

```python
kmeans = KMeans(n_clusters=3, init="k-means++")
```

```python
kmeans.fit_predict(X_train)
```

```python
y_pred = kmeans.predict(X_train)
```

```python
y_test_pred = kmeans.predict(X_test)
plt.scatter(X_test[:, 0], X_test[:, 1], c=y_test_pred, cmap='viridis')
plt.title('K-Means Clusters on Test Data')
plt.xlabel('Feature 1 (Scaled)')
plt.ylabel('Feature 2 (Scaled)')
plt.colorbar(label='Cluster Label')
plt.show()
```

```python
from sklearn.linear_model import LogisticRegression
lr = LogisticRegression()
```

```python
lr.fit(X_train, y_train)
```

```python
y_pred_lr = lr.predict(X_test)
```

The **silhouette coefficient** is a measure of how similar an object is to its own cluster (cohesion) compared to other clusters (separation). It ranges from -1 to 1:
-   A score close to +1 indicates that the object is well matched to its own cluster and poorly matched to neighboring clusters.
-   A score close to 0 indicates overlapping clusters.
-   A score close to -1 indicates that the object might have been assigned to the wrong cluster.
