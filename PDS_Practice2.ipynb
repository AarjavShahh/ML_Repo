```python
import matplotlib.pyplot as plt
```

```python
import numpy as np
```

```python
cat = ['a', 'b', 'c', 'd']
val1 = [1, 2, 3, 4]
val2 = [5, 6, 7, 8]

x = np.arange(len(cat))

plt.figure(figsize=(10, 6))
plt.bar(x, val1, color="blue", width=0.6)
plt.xticks(x, cat)
plt.title("Column Chart")
plt.xlabel("Categories")
plt.ylabel("Values")
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.barh(cat, val1, color="orange")
plt.title("Horizontal Bar Chart")
plt.xlabel("Categories")
plt.ylabel("Values")
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.bar(x, val1, color="green", label="Value1")
plt.bar(x, val2, bottom=val1, color="red", label="Value2")
plt.xticks(x, cat)
plt.title("Stacked Bar Chart")
plt.xlabel("Categories")
plt.ylabel("Values")
plt.legend()
plt.show()
```

* plt.legend() displays that categorical difference between values1 and values2 on the top left corner.

```python
x = np.arange(1, 6)
y1 = np.array([2, 4, 6, 5, 7])
y2 = np.array([1, 3, 5, 4, 6])

plt.figure(figsize=(10, 6))
plt.stackplot(x, y1, y2, labels=["Series1", "Series2"], alpha=0.9)
plt.title("Area Chart")
plt.xlabel("X-axis")
plt.ylabel("Values")
plt.legend()
plt.show()
```

* alpha is used for the opacity of the chart.

```python
fig, ax1 = plt.subplots(figsize=(8, 5))
ax2 = ax1.twinx()
ax1.bar(x, y1, color="green", label="Bar Data")
ax2.plot(x, y2, color="red", marker="o", label="Line Data")
ax1.set_xlabel("X-axis")
ax1.set_ylabel("Bar Values")
ax2.set_ylabel("Line Values")
plt.title("Dual Axis Chart")
plt.show()
```

```python
widths = [0.5, 1.0, 1.5, 2.0, 2.5]
val = [9, 4, 2, 6, 4]

lefts = np.cumsum([0] + widths[:-1])
plt.figure(figsize=(8, 6))
plt.bar(lefts, val, width=widths, color="lightgreen", edgecolor="black")
plt.title("Meeko Chart")
plt.xlabel("Categories")
plt.ylabel("Values")
plt.show()
```
