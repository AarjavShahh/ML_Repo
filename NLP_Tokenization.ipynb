# Poter Stemmer


```python
from nltk.stem import PorterStemmer
stemming = PorterStemmer()
```

```python
for i in words:
  print(i, "--->", stemming.stem(i))
```

# Regular expression stemmer class

It takes a single regular expression and removes any prefix and suffix that matches the expression.

```python
from nltk.stem import RegexpStemmer
stemming = RegexpStemmer('ing$|s$|e$|able$', min=4) ## Means wherever there is s, e, ing and able, remove those characters.
stemming.stem('eating')
```

```python
stemming = RegexpStemmer('ing|s$|e$|able$', min=4)
stemming.stem('ingeating')
```

```python
stemming = RegexpStemmer('ing|s$|e$|able$|es$', min=4)
stemming.stem('goes')
```

# Lemmatization

It is a technique just like stemming. The output we will get after lemmatization is called Lemma which is the root word rather than root stem. After lemmatization we will be getting a valid word that means the same thing.

```python
from nltk.stem import PorterStemmer
from nltk.corpus import stopwords
```

```python
import nltk
nltk.download('stopwords')
```

```python
stopwords.words('english')
```

```python
stemmer = PorterStemmer()
```

```python
import nltk
nltk.download('punkt_tab')
nltk.download('punkt')

```

```python
sentences = nltk.sent_tokenize(paragraph)
```

```python
sentences
```

```python
type(sentences)
```

```python
## Apply stopwords and filter and then apply stemming

for i in range(len(sentences)):
  words = nltk.word_tokenize(sentences[i])
  words = [stemmer.stem(word) for word in words if word not in set(stopwords.words('english'))] ## This will apply stemming if stopwords are present in the paragraph.
  sentences[i] = ' '.join(words) ## Converting all the words into sentences
```

```python
sentences
```

```python
from nltk.stem import SnowballStemmer
snowballstemmer = SnowballStemmer('english')
```

```python
## Apply stopwords and filter and then apply snowballstemming

for i in range(len(sentences)):
  words = nltk.word_tokenize(sentences[i])
  words = [snowballstemmer.stem(word) for word in words if word not in set(stopwords.words('english'))] ## This will apply stemming if stopwords are present in the paragraph.
  sentences[i] = ' '.join(words) ## Converting all the words into sentences
```

```python
sentences
```

```python
from nltk.stem import WordNetLemmatizer
lemmatizer = WordNetLemmatizer()
```

```python
## Apply stopwords and filter and then apply lemmatization

for i in range(len(sentences)):

  words = nltk.word_tokenize(sentences[i])
  words = [lemmatizer.lemmatize(word.lower(), pos='v') for word in words if word not in set(stopwords.words('english'))] ## This will apply stemming if stopwords are present in the paragraph.
  sentences[i] = ' '.join(words) ## Converting all the words into sentences
```
